```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

# Instruction: the function toWordCloud take in input a dictionary and (if you want ) the path of a png image and outputs the wordcloud both in the notebook and as image in your working directory

```python
def transform_format(val):
    if val == 0:
        return 255
    else:
        return val
    
def toWordCloud(dictionary,image_path=None):
    if image_path==None: 
        wc = WordCloud(background_color="white", max_words=1000 #mask=transformed_wine_mask
               , contour_width=3, contour_color='firebrick')
        wc.generate_from_frequencies(dictionary)

        wc.to_file("output.png")

        plt.figure(figsize=[20,10])
        plt.imshow(wc, interpolation='bilinear')
        plt.axis("off")
        plt.show()
    else:
        wine_mask = np.array(Image.open(image_path))
        #wine_mask
    
        transformed_wine_mask = np.ndarray((wine_mask.shape[0],wine_mask.shape[1]), np.int32)
        for i in range(len(wine_mask)):
            transformed_wine_mask[i] = list(map(transform_format, wine_mask[i]))
        
        wc = WordCloud(background_color="white", max_words=1000, mask=transformed_wine_mask
               , contour_width=3, contour_color='firebrick')
        wc.generate_from_frequencies(dictionary)

        wc.to_file("output.png")

        plt.figure(figsize=[20,10])
        plt.imshow(wc, interpolation='bilinear')
        plt.axis("off")
        plt.show()
```

```python
dictionary= {u'asielzoekers': 0.0034861170591325486,
 u'belastingverlaging': 0.0018551991553514675,
 u'buma': 0.0020712555982839408,
 u'islam': 0.0029519544163739155,
 u'moslims': 0.0027958002747301355,
 u'ouderen': 0.0022803123245457566,
 u'pechtold': 0.0021525864470786928,
 u'president': 0.003281844532743345,
 u'rutte': 0.0023488684001475584,
 u'samsom': 0.0019304632325980841} 

image_path= "winepattern.png"
```
